```python
# zmiana waluty na USD (1.08.2022)
for i in range(len(data['Average Cost for two'])):
    if data['Currency'][i] == 'Botswana Pula(P)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 0.080
        
    elif data['Currency'][i] == 'Brazilian Real(R$)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 0.1933
        
    elif data['Currency'][i] == 'Emirati Diram(AED)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 0.27
        
    elif data['Currency'][i] == 'Indian Rupees(Rs.)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 0.013
        
    elif data['Currency'][i] == 'Indonesian Rupiah(IDR)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 0.000067
        
    elif data['Currency'][i] == 'NewZealand($)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 0.628821
    
    elif data['Currency'][i] == 'Pounds(\x8c£)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 1.22587
        
    elif data['Currency'][i] == 'Qatari Rial(QR)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 0.27
        
    elif data['Currency'][i] == 'Rand(R)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 0.061
        
    elif data['Currency'][i] == 'Sri Lankan Rupee(LKR)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 0.0028
        
    elif data['Currency'][i] == 'Turkish Lira(TL)':
        data['Average Cost for two'][i] = data['Average Cost for two'][i] * 0.056
    
```

```python
data['Has Table booking'].replace({'No': 0, 'Yes':1}, inplace=True)
data['Is delivering now'].replace({'No': 0, 'Yes':1}, inplace=True)
data['Switch to order menu'].replace({'No': 0, 'Yes':1}, inplace=True)
data['Has Online delivery'].replace({'No': 0, 'Yes':1}, inplace=True)
```

```python
data = data.drop(['Restaurant Name', 'Address', 'Locality', 'Locality Verbose', 'Cuisines', 'Rating color'], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
data
```

```python
# dzielenie danych na zależne i niezależne oraz budowanie modelu

x = data.drop('Aggregate rating', axis=1)
y = data['Aggregate rating']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import StackingRegressor, ExtraTreesRegressor

lr = LinearRegression()
svr = SVR()
rfr = RandomForestRegressor()
dtr = DecisionTreeRegressor()

clf = [('lr', lr), ('svr', svr), ('rfr', rfr), ('dtr', dtr)]
```

```python
ran = RandomForestRegressor()

stack_model = StackingRegressor(estimators=clf, final_estimator=ran)
```

```python
stack_model.fit(x_train, y_train)
```

```python
stack_model.score(x_train, y_train)
```

```python
stack_model.score(x_test, y_test)
```

```python
etr = ExtraTreesRegressor()
```

```python
etr.fit(x_train, y_train)
```

```python
etr.score(x_train, y_train)
```

```python
# overfitting
etr.score(x_test, y_test)
```

```python
import joblib

joblib.dump(stack_model, 'Rating_Model.pkl')
```
